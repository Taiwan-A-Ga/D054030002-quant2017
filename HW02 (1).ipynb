```python
# 載入模組
import os

# 這是main()的註解
def main():
    # Python 3 需使用()框住字串
    print ('Hello world!')
    # 若有''的存在則須使用""包住字串
    print ("This is Alice's greeting.")
    # 如果先加了\就不會影響''的使用
    print ('This is Bob\'s greeting.')
    
    # 使用Function
    foo(5, 10)
    
    # 字串可以用數字重複
    print ('=' * 10)
    # 找出目前所在目錄
    print ('Current working directory is ' + os.getcwd())
    
    # 變數需先事先定義
    counter = 0
    # 由原初始值以+1的方式不斷更新
    counter += 1
    
    # 建立list的方法，list裡可以包含多種不同Type的值，甚至是另一個list
    food = ['apples', 'oranges', 'cats']
    
    # 把food當作迴圈變數，程式會自動針對每個值個別做一次迴圈
    for i in food:
        print ('I like to eat ' + i)
    
    # 使用range function列出0~9的數值
    print ('Count to ten')
    for i in range(10):
        print (i)
   
 # 自定義foo function
def foo(param1, secondParam):
    # 自定義公式
    res = param1 + secondParam
    
    # 在字串裡插入Function裡算出來的值
    print ('%s plus %s is equal to %s' % (param1, secondParam, res))
    
    # 若res算出來的值小於50，則印出foo
    if res < 50:
        print('foo')
    
    # 若res算出來的值大於等於50 且 param1等於42或secondParam等於24，則印出bar
    elif (res >= 50) and ((param1 == 42) or (secondParam == 24)):
        print('bar')
        
    # 出現除了上述以外的條件以外則印出moo
    else:
        print('moo')
        
    return res # This is a one-line comment.
    # 只有一開始的indendation會影響，後面的字串不影響
    '''A multi-
line string, but can also be a multi-line comment.'''
    
# indendation的結束
# 使用這樣的寫法可以製造該程式被'執行'及被'import'兩種不同的結果，若Value顯示__main__表示該程式被'執行'
# 而不是被import，若該程式是被'import'則該程式不會被執行
if __name__ == '__main__':
    main()
    
```
